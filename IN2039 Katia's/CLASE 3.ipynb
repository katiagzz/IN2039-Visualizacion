```python
import numpy as np
import pandas as pd
df = pd.read_csv('datasets/census.csv')
#pd.set_option('display.max_columns',500, 'display.max_rows', 500)
df.head()
```

```python
pd.pivot_table(df,index='STNAME')
```

```python
pd.pivot_table(df, index='STNAME',values='BIRTHS2010') #default is : aggfunc='mean'
```

```python
pd.pivot_table(df,index='STNAME', values='BIRTHS2010',aggfunc = 'sum')
```

```python
pd.pivot_table(df,index='STNAME', values='BIRTHS2010',aggfunc = 'median')
```

```python
pd.pivot_table(df,index='STNAME', values='BIRTHS2010',aggfunc = 'mean')
```

```python
pd.pivot_table(df,index='STNAME', values='BIRTHS2010',aggfunc = 'count',margins = True) 
#margins gives the 'ALL' at the end and botom
```

```python
pd.pivot_table(df,index='STNAME', values='BIRTHS2010',aggfunc = 'count')
```

```python
df = pd.read_csv('datasets/airline_data.csv')
df.head(500)
```

```python
pd.pivot_table(df, 
               index=['Year', 'Quarter'], 
               values='DepDelay', 
               columns='DepartureDelayGroups', 
               aggfunc= 'sum',
               margins=True
              ) #margins gives the 'ALL' at the end and botom
```

```python

pd.pivot_table(df, 
               index=['Year', 'Quarter'], 
               values='AirTime', 
               aggfunc='sum', 
               margins=True
              ) #margins gives the 'ALL' at the end and botom
```

```python
data_file = 'datasets/sample_pivot.csv'
data = pd.read_csv(data_file)
data.head()
```

```python
pd.pivot_table(data, 
               index='Region'
              )
```

```python
average_sales = pd.pivot_table(data, 
                               index='Region', 
                               values='Sales'
                              )
average_sales
```

```python
total_sales_pivot = pd.pivot_table(data, 
                                   index='Region',
                                   values='Sales', 
                                   aggfunc='sum'
                                  )
total_sales_pivot
```

```python
sales_pivot = pd.pivot_table(data, 
                             index='Region', 
                             values='Sales', 
                             aggfunc=['sum', 'mean','median']
                            )
sales_pivot
```

```python
pd.pivot_table(data, 
               index='Region', 
               values='Units', 
               columns='Type'
              )
```

```python
pd.pivot_table(data, 
               index='Region', 
               values='Units', 
               columns='Type', 
               fill_value=0
              )
```

```python
pd.pivot_table(data, 
               index='Region', 
               values='Sales', 
               columns='Type', 
               aggfunc='sum', 
               margins='sum', 
               margins_name='Total'
              )
```
