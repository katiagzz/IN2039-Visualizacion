```python
path = 'datasets/Cartwheeldata.csv'
df = pd.read_csv(path)
df.head(2)
```

```python
sns.lmplot(x ='Wingspan',
          y = 'CWDistance',
          data = df,
          fit_reg = False, 
           hue = 'Gender')
# fit_reg False sin linea de regresion y True (Default) para poner la linea de regresion 

plt.title('Scatter plot of Wingspan vs CWDistance')
plt.show()
```

```python
sns.swarmplot(x='Gender',
            y = 'CWDistance',
            data = df
            )
plt.title('Gender vs CWDistance')
plt.show()
```

```python
sns.boxplot(data = df.loc[df['Gender'] == 'M',
           ['Height', 'Wingspan', 'CWDistance']])
plt.title('Box Plot')
plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(1,2,1) #(renglon, columnas, posicion)
sns.boxplot(data= df.loc[df['Gender'] =='M', 
                         ['Score']])

ax2 =fig.add_subplot(1,2,2)
sns.boxplot(data= df.loc[df['Gender'] =='F', 
                         ['Score']])

ax1.title.set_text('Male Score')
ax2.title.set_text('Female Score')
plt.show()
```

```python
sns.distplot(df['CWDistance'])
plt.show()
```

```python
sns.countplot(x='Gender', 
              data =df)
plt.xticks (rotation = 45)
plt.yticks (rotation = 45)
plt.show()
```

```python
df = pd.read_excel('datasets/Canada.xlsx', 
                  sheet_name = 'Canada by Citizenship', 
                  skiprows = range(20),
                  skipfooter  = 2,
                   engine = 'openpyxl'
                  )
df.head(:)
```

```python
df['Total'] = df.iloc[:,9:].sum(axis=1)
df.head()
```

```python
df.columns
```

```python
df.columns= list(map(str, df.columns))
years = list(map(str, range(1980, 2014)))
df.head(2)
years 
```
