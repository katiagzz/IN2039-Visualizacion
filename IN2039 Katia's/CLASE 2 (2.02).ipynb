SUMA DE TODOS LOS ELEMENTOS DE UNA MATRIZ

```python
x = np.array([[1,2],[3,4]])

### Compute sum of all elements; prints "10"
print(np.sum(x))
```

SUMA DE LAS COLUMNAS DE UNA MATRIZ

```python
### Compute sum of each column; prints "[4 6]"
print(np.sum(x, axis=0))
```

SUMA DE LOS RENGLONES DE UNA MATRIZ

```python
### Compute sum of each row; prints "[3 7]"
print(np.sum(x, axis=1))
```

```python
x = np.array([[1,2],[3,4]])

### Compute mean of all elements; prints "2.5"
print(np.mean(x))

### Compute mean of each column; prints "[2 3]"
print(np.mean(x, axis=0)) 

### Compute mean of each row; prints "[1.5 3.5]"
print(np.mean(x, axis=1))
```

NORMAL GAUSSIAN DISTRIBUTION

```python
# * importing all ( = import pylab)
from pylab import * 

# Create some test data
dx = .01

X  = np.arange(-2,2,dx)
Y  = exp(-X**2)

# Normalize the data to a proper PDF (probability density funcion )
Y /= (dx*Y).sum()

# Compute the CDF (cumulative distrubution funcion)
CY = np.cumsum(Y*dx)

# Plot both
plot(X,Y)


show()
```

```python
plot(X,CY,'r--')
show()
```

```python
plot(X,Y, 'y')
plot(X, CY, 'g--')
show()
```

```python
### Compute the Normal CDF of certain values.
print(stats.norm.cdf(np.array([-2, -1., -0.5, 0, 1, 3, 4, 2, 6])))
```
